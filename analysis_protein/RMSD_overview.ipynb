```python
import MDAnalysis as mda
import nglview as nv
from MDAnalysis.analysis import rms, distances, align, pca, rms
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle
```

```python
# Load trajectories (.xtc) and topology file (.tpr)
trajfiles1= ("/scratch/gent/vo/000/gvo00003/vsc48847/apo_protein/charmm-gui-7064934746/gromacs/Rep1/step7_production_rep1_final.xtc", 
             "/scratch/gent/vo/000/gvo00003/vsc48847/apo_protein/charmm-gui-7064934746/gromacs/step7_production_rep2_final.xtc",
             "/scratch/gent/vo/000/gvo00003/vsc48847/apo_protein/charmm-gui-7064934746/gromacs/step7_production_rep3_final.xtc",
            )
topfile1 = ("/scratch/gent/vo/000/gvo00003/vsc48847/apo_protein/charmm-gui-7064934746/gromacs/Rep1/step7_production_rep1.tpr")
trajfiles2 = ("/scratch/gent/vo/000/gvo00003/vsc48847/BDN_bound/gromacs/step7_production_rep1_f.xtc",
        "/scratch/gent/vo/000/gvo00003/vsc48847/BDN_bound/gromacs/step7_production_rep2_f.xtc",
        "/scratch/gent/vo/000/gvo00003/vsc48847/BDN_bound/gromacs/step7_production_rep3_f.xtc",
            )
topfile2 = ("/scratch/gent/vo/000/gvo00003/vsc48847/BDN_bound/gromacs/step7_production_rep1.tpr")
trajfiles3 = ("/scratch/gent/vo/000/gvo00003/vsc48847/BDQ_bound/gromacs/step7_production_rep1_f.xtc",
        "/scratch/gent/vo/000/gvo00003/vsc48847/BDQ_bound/gromacs/step7_production_rep2_f.xtc",
        "/scratch/gent/vo/000/gvo00003/vsc48847/BDQ_bound/gromacs/step7_production_rep3_f.xtc",
            )
topfile3 = ("/scratch/gent/vo/000/gvo00003/vsc48847/BDQ_bound/gromacs/step7_production_rep1.tpr")
apo1 = mda.Universe(topfile1, trajfiles1[0])
apo2 = mda.Universe(topfile1, trajfiles1[1])
apo3 = mda.Universe(topfile1, trajfiles1[2])
apo = mda.Universe(topfile1, trajfiles1)

bdn1 = mda.Universe(topfile2, trajfiles2[0])
bdn2 = mda.Universe(topfile2, trajfiles2[1])
bdn3 = mda.Universe(topfile2, trajfiles2[2])
bdn = mda.Universe(topfile2, trajfiles2)

bdq1 = mda.Universe(topfile3, trajfiles3[0])
bdq2 = mda.Universe(topfile3, trajfiles3[1])
bdq3 = mda.Universe(topfile3, trajfiles3[2])
bdq = mda.Universe(topfile3, trajfiles3)
```

```python
import matplotlib.pyplot as plt
rmsd_apo1 = rms.RMSD(apo1, select = "protein") 
rmsd_apo2= rms.RMSD(apo2, select = "protein") 
rmsd_apo3= rms.RMSD(apo3, select = "protein")

rmsd_bdn1 = rms.RMSD(bdn1, select = "protein") 
rmsd_bdn2= rms.RMSD(bdn2, select = "protein") 
rmsd_bdn3= rms.RMSD(bdn3, select = "protein") 

rmsd_bdq1 = rms.RMSD(bdq1, select = "protein") 
rmsd_bdq2= rms.RMSD(bdq2, select = "protein") 
rmsd_bdq3= rms.RMSD(bdq3, select = "protein") 
# run the calculation 
rmsd_apo1.run() 
rmsd_apo2.run() 
rmsd_apo3.run() # access results 

rmsd_bdn1.run() 
rmsd_bdn2.run() 
rmsd_bdn3.run() # access results 

rmsd_bdq1.run() 
rmsd_bdq2.run() 
rmsd_bdq3.run() # access results 

rmsd_data_apo1 = rmsd_apo1.results.rmsd 
rmsd_data_apo2 = rmsd_apo2.results.rmsd 
rmsd_data_apo3 = rmsd_apo3.results.rmsd

rmsd_data_bdn1 = rmsd_bdn1.results.rmsd 
rmsd_data_bdn2 = rmsd_bdn2.results.rmsd 
rmsd_data_bdn3 = rmsd_bdn3.results.rmsd

rmsd_data_bdq1 = rmsd_bdq1.results.rmsd 
rmsd_data_bdq2 = rmsd_bdq2.results.rmsd 
rmsd_data_bdq3 = rmsd_bdq3.results.rmsd
# Entire protein
fig, ax = plt.subplots()

# ---------------- APO (blauw tinten)
ax.plot(rmsd_data_apo1[:,1]/1000, rmsd_data_apo1[:,2],
        label="apo-1", color="#c6dbef")
ax.plot(rmsd_data_apo2[:,1]/1000, rmsd_data_apo2[:,2],
        label="apo-2", color="#9ecae1")
ax.plot(rmsd_data_apo3[:,1]/1000, rmsd_data_apo3[:,2],
        label="apo-3", color="#6baed6")

# ---------------- BDN (groen tinten)
ax.plot(rmsd_data_bdn1[:,1]/1000, rmsd_data_bdn1[:,2],
        label="bdn-1", color="#c7e9c0")
ax.plot(rmsd_data_bdn2[:,1]/1000, rmsd_data_bdn2[:,2],
        label="bdn-2", color="#74c476")
ax.plot(rmsd_data_bdn3[:,1]/1000, rmsd_data_bdn3[:,2],
        label="bdn-3", color="#31a354")

# ---------------- BDQ (rood tinten)
ax.plot(rmsd_data_bdq1[:,1]/1000, rmsd_data_bdq1[:,2],
        label="bdp-1", color="#fcbba1")
ax.plot(rmsd_data_bdq2[:,1]/1000, rmsd_data_bdq2[:,2],
        label="bdp-2", color="#fb6a4a")
ax.plot(rmsd_data_bdq3[:,1]/1000, rmsd_data_bdq3[:,2],
        label="bdp-3", color="#cb181d")

# labels & styling
ax.set_xlabel("Time (ns)")
ax.set_ylabel("RMSD (Å)")
# y-as van 0 tot 7 Å
ax.set_ylim(0, 7)
ax.legend(ncol=3, fontsize=8)
plt.tight_layout()

plt.savefig("rmsd_plot_all_replicas_bis.png", dpi=300, bbox_inches='tight')
plt.show()
```
